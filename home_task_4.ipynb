с помощью библиотеки folium нанесем геоданные на реальную карту

```python
pip install folium
```

```python
import folium
this_map = folium.Map(prefer_canvas=True)

def plotDot(point):
    folium.CircleMarker(
        location=[point.lat, point.long],
        radius=2,
        popup=point.price
    ).add_to(this_map)
    
df.apply(plotDot, axis=1)

this_map.fit_bounds(this_map.get_bounds())

this_map
```

Для оценки географического расположения домов эта карта полезна, но не хватает отображения стоимости домов

Сделать это можно с помощью создания тепловой карты c весами с использованием той же библиотеки Folium (и с помощью неравнодушных кодеров с сайта stackoverflow)

```python
from folium import plugins
from folium.plugins import HeatMap

# подготовим данные: координаты и стоимость
heat_price = df[["lat","long","price"]]
map_price = folium.Map(location=[df['lat'].mean(), df['long'].mean()], zoom_start=9 )
heat_price
```

```python
heat_price = heat_price.values.tolist()
heat_price
```

```python
HeatMap(heat_price,radius=13).add_to(map_price)
map_price
```

Данная карта довольно наглядна, но необходимо учесть, что она скорее показывает **количество** домов в районе, а не среднюю стоимость


Интересно будет представить таким же образом состояние домов ('grade') на карте

Но сделать это для наглядности необходимо с разбиением на определенные категории: допустим, дома делятся на дома в плохом состоянии (grade <=5), дома в хорошем состоянии (5<grade<=10), дома в отличном состоянии (grade>=11)



```python
heat_grade = df[df['grade']<=5][["lat","long","grade"]]
map_grade = folium.Map(location=[df['lat'].mean(), df['long'].mean()], zoom_start=9)
HeatMap(heat_grade,radius=15).add_to(map_grade)
map_grade
```

На данной карте выделены районы c домами в плохом состоянии

```python
heat_grade = df[(df['grade']<=10) & (df['grade']>5)][["lat","long","grade"]]
map_grade = folium.Map(location=[df['lat'].mean(), df['long'].mean()], zoom_start=9)
HeatMap(heat_grade,radius=15).add_to(map_grade)
map_grade
```

На данной карте выделены районы c домами в среднем состоянии (их больше и по количеству), дома в таком состоянии представлены практически повсеместно

```python
heat_grade = df[(df['grade']>10)][["lat","long","grade"]]
map_grade = folium.Map(location=[df['lat'].mean(), df['long'].mean()], zoom_start=9)
HeatMap(heat_grade,radius=15).add_to(map_grade)
map_grade
```

Выделены районы с самыми лучшими домам, характерно выделяется набережная, также при увеличениии масштаба можно выделить престижные микрорайоны

Выводы: в ходе оценки взаимосвязи признаков и стоимости дома установлено, что наиболее значимыми признаками, влияющими на стоимость дома, являются
    
    1. Состояние дома (grade)
    2. Жилая площадь
    3. Площадь чердака
    
    Также немалую роль играет расположение дома, а именно: близость к водоему, престижность района (микрорайона)
